# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```
print(x)
```

```
print(y)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
# finds the columns to apply changes and calculates the mean of that column
imputer.fit(x[:,1:3])
# changing the affected rows i.e filling the nan values with mean (only column 2 and 3 will change)
x[:,1:3]=imputer.transform(x[:,1:3])
```

```
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])],remainder='passthrough')
# x should be a numpy array
x= np.array(ct.fit_transform(x))
```

```
print(x)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
```

```
print(y)
```

## Feature Scaling

```
# This step is necessary because it might happen that one feature(having larger value) might dominate the one with lower value.
# hence, bringing them on a common scale is necessary
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# connecting to matrix of features and transforming
x=sc.fit_transform(x)
```

```
print(x)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```
print(x_test)
```
